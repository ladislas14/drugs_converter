```python
import psycopg2
import json
from lxml import etree, html

from dotenv import load_dotenv
load_dotenv()

conn =  psycopg2.connect('')
conn.autocommit = True
```

```python
records = []
with conn.cursor() as cur:
    cur.execute('SELECT id, html FROM inn.inns_html')
    records = cur.fetchall()
```

```python
insert_sql = '''INSERT INTO inn.inns_json (id, translations)
                VALUES %s
                ON CONFLICT (id)
                DO UPDATE
                SET id = excluded.id,
                    translations = excluded.translations,
                    updated_at = excluded.updated_at
'''
```

```python
from psycopg2.extras import execute_values

translations = []
for record in records:
    print('Treating record' + str(record[0]))
    doc = html.fromstring(record[1])

    # Parsing substance names
    languages_table = doc.cssselect('#ctl00_rightcolumn_innForm')[0]
    languages = {
        'English': languages_table.cssselect('h2')[0].text_content().strip()
    }
    for tr in languages_table.cssselect('tr table > tr'):
        #print(etree.tostring(tr, encoding='unicode', pretty_print=True))
        if (len(tr.cssselect('td:nth-child(1) > label')) == 0): ## S'il n'y a pas de label
            continue
        
        if (len(tr.cssselect('td:nth-child(2) > span')) == 0): ## S'il n'y a pas de valeur
            continue
            
            
        language = tr.cssselect('td:nth-child(1) > label')[0].text_content().strip()
        name_in_language = tr.cssselect('td:nth-child(2) > span')[0].text_content().strip()
        languages[language] = name_in_language
    
    #Parsing ATC Codes
    atc_codes_table = languages_table.cssselect('tr table > tr td table')
    if(len(atc_codes_table) == 0):
        languages['ATC Codes'] = {}
    else:
        atc_codes = []
        for atc_code in atc_codes_table[0].cssselect('tr > td > font'):
            atc_codes.append(atc_code.text_content().strip())
        languages['ATC Codes'] = atc_codes


    # Parsing alternates names (i.e. names used by national commissions) and national commisions using same name
    tables = doc.cssselect('#ctl00_rightcolumn_box .c > table')
    if(len(tables) > 1):
        for data_table in tables[1::]:
            data_label = data_table.cssselect('tr:first-child td:first-child b')[0].text_content().strip()

            if (data_label == "National commission(s) using same name"): ## National commission(s) using same name
                data = []
                for tr in data_table.cssselect('tr')[1::]:
                    data.append(tr.cssselect('td:nth-child(1) > font')[0].text_content().strip())
                languages[data_label] = data
            else: ## Alternate names
                data = {}
                for tr in data_table.cssselect('tr')[1::]:
                    commission_name = tr.cssselect('td:nth-child(1) > font')[0].text_content().strip()
                    alternate_name = tr.cssselect('td:nth-child(2) > font')[0].text_content().strip()
                    data[commission_name] = alternate_name
                languages[data_label] = data
    
    translations.append((record[0], json.dumps(languages)))

with conn.cursor() as cursor2:
        #cursor2.execute(insert_sql, (record[0], json.dumps(languages)))
        execute_values(cursor2,
        insert_sql,
        translations)
```
