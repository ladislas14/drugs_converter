```python
records = []
with conn.cursor() as cur:
    cur.execute('SELECT id, html FROM inn.inns_html LIMIT 100')
    records = cur.fetchall()
```

```python
insert_sql = '''INSERT INTO inn.inns_json (id, created_at, updated_at, translations)
        VALUES (%s, now(), now(), %s)
        ON CONFLICT (id)
        DO UPDATE SET
            (updated_at, translations)
            = (EXCLUDED.updated_at, EXCLUDED.translations) ;
'''
```

```python
for record in records:
    print('Treating record' + str(record[0]))
    doc = html.fromstring(record[1])
    languages_table = doc.cssselect('#ctl00_rightcolumn_innForm')[0]
    languages = {
        'English': languages_table.cssselect('h2')[0].text_content().strip()
    }
    for tr in languages_table.cssselect('tr table > tr'):
        #print(etree.tostring(tr, encoding='unicode', pretty_print=True))
        if (len(tr.cssselect('td:nth-child(1) > label')) == 0): ## S'il n'y a pas de label
            continue
        
        if (len(tr.cssselect('td:nth-child(2) > span')) == 0): ## S'il n'y a pas de valuer
            continue
            
            
        language = tr.cssselect('td:nth-child(1) > label')[0].text_content().strip()
        name_in_language = tr.cssselect('td:nth-child(2) > span')[0].text_content().strip()
        languages[language] = name_in_language
    
    with conn.cursor() as cursor2:
        cursor2.execute(insert_sql, (record[0], json.dumps(languages)))
```
